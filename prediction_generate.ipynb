```python
import pandas as pd
result=pd.read_csv('C:\\Users\\fcn\\Desktop\\dataset\\dataset\\prediction.csv',encoding='utf-8',header=None,names=['第1列','第2列','第3列','第4列','第5列','第6列','第7列','第8列','第9列','第10列','第11列','第12列','第13列','第14列','第15列',])
result.info()
```

```python
def dayudengyuling(x):
    if(x>=0):
        return True;
    else:
        return False;
result=result[result['第1列'].apply(dayudengyuling)&result['第2列'].apply(dayudengyuling)&result['第3列'].apply(dayudengyuling)&result['第4列'].apply(dayudengyuling)&result['第5列'].apply(dayudengyuling)&result['第6列'].apply(dayudengyuling)&result['第7列'].apply(dayudengyuling)&result['第8列'].apply(dayudengyuling)&result['第9列'].apply(dayudengyuling)&result['第10列'].apply(dayudengyuling)&result['第11列'].apply(dayudengyuling)&result['第12列'].apply(dayudengyuling)&result['第13列'].apply(dayudengyuling)&result['第14列'].apply(dayudengyuling)&result['第15列'].apply(dayudengyuling)]
result
```

```python
def sishewuru(x):
    return int(round(x))
result['第1列']=result['第1列'].apply(sishewuru)
result['第2列']=result['第2列'].apply(sishewuru)
result['第3列']=result['第3列'].apply(sishewuru)
result['第4列']=result['第4列'].apply(sishewuru)
result['第5列']=result['第5列'].apply(sishewuru)
result['第6列']=result['第6列'].apply(sishewuru)
result['第7列']=result['第7列'].apply(sishewuru)
result['第8列']=result['第8列'].apply(sishewuru)
result['第9列']=result['第9列'].apply(sishewuru)
result['第10列']=result['第10列'].apply(sishewuru)
result['第11列']=result['第11列'].apply(sishewuru)
result['第12列']=result['第12列'].apply(sishewuru)
result['第13列']=result['第13列'].apply(sishewuru)
result['第14列']=result['第14列'].apply(sishewuru)
result['第15列']=result['第15列'].apply(sishewuru)
result
```
