```python
import pandas as pd
import numpy as np
pay_info=pd.read_csv('C:\\Users\\fcn\\Desktop\\dataset\\dataset\\user_pay.txt',names=['用户id','商家id','支付时间'])
shop_info=pd.read_csv('C:\\Users\\fcn\\Desktop\\dataset\\dataset\\shop_info.txt',names=['商家id','市名','所在位置编号','人均消费','评分','评论数','门店等级','一级品类名称','二级分类名称','三级分类名称'])
pay_info
```

```python
pay_info['支付时间']=pay_info['支付时间'].str.slice(start=0,stop=10)
pay_info
```

```python
pay_info=pay_info.groupby(['商家id','支付时间']).size()
pay_info
```

```python
pay_info=pay_info.reset_index()
pay_info
```

```python
pay_info.rename(columns={0:'客户流量'},inplace=True)
pay_info
```

```python
merge_info=pay_info.merge(shop_info,on=['商家id'])
merge_info
```

```python
merge_info.to_csv('C:\\Users\\fcn\\Desktop\\dataset\\dataset\\merge_info.txt',index=False)
```
